```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import astropy
%matplotlib inline
import pylab
import scipy.interpolate as sp
from numpy import array
```

```python
Lambda= array([447.15,471.31,492.19,501.57,587.56,667.81,706.52])
n = array([1.8206,1.8120,1.8056,1.8028,1.7851,1.7751,1.7713])
D =  array([70.82,69.65,69.15,68.80,66.15,64.90,64.43])
```

```python
more_n = np.linspace(1.8206,1.7713,num=50)
more_Lambda = np.linspace(447.15,706.52,num=50)
```

```python
csfont = {'fontname':'Cambria Math'}

f = plt.figure(figsize=(10,5))
plt.plot(Lambda,n,lw=2,c='blue')
plt.xlabel(r'$\lambda$ [nm]',fontsize = 15,**csfont)
plt.scatter(Lambda,n,s=70,marker='o',c='red')
plt.ylabel('n',fontsize = 15,**csfont)
plt.title('Index of Refraction vs. Wavelength',fontsize = 20,**csfont)
plt.show()
```

```python
f = plt.figure(figsize=(10,5))
plt.plot(Lambda,D,lw=2,c='blue')
plt.scatter(Lambda,D,s=70,marker='o',c='red')
plt.xlabel(r'$\lambda$ [nm]',fontsize = 15,**csfont)
plt.ylabel(r'$\theta$' ,fontsize = 15,**csfont)
plt.title('Angle of Deviation vs. Wavelength',fontsize = 20,**csfont)
plt.show()
```
